```python
import sys
def problemA():
    linec = 0
    for line in open("inputA.txt"):
        s = line.split()
        if linec == 0:
            linec += 1
            continue
        if linec % 2:
            n = int(s[0])
            linec += 1
            continue
        else:
            word = s[0]
            from_left = []
            from_left_c = 0
            from_right = []
            from_right_c = 0
            for c in word:
                from_left.append(from_left_c)
                if c == "R": from_left_c += 1
            for c in word[::-1]:
                from_right.append(from_right_c)
                if c == "L": from_right_c += 1
            from_right = from_right[::-1]
            total = [from_left[i] + from_right[i] for i in range(len(from_right))]
            num0s = len([el for el in total if el == 0])
            if num0s > 2:
                print(-1)
                continue
            linec += 1
            printed = False
            if num0s == 0:
                print(0)
                continue
            if num0s == 1:
                for i in range(n - 1):
                    if total[i] == 0 and (word[i], word[i + 1]) == ("L", "R"):
                        print(i + 1)
                        printed = True
                        break
            if num0s == 2:
                for i in range(n - 1):
                    if total[i] + total[i + 1] == 0 and (word[i], word[i + 1]) == ("L", "R"):
                        print(i + 1)
                        printed = True
                        break
            if not printed:
                print(-1)
problemA()
```

```python
def problemB():
    linec = 0
    fot = {
        1: 0,
        2: 1,
        3: 0,
        4: 2,
        5: 0,
        6: 1,
        7: 0,
        8: 3,
        9: 0,
        10: 1,
        11: 0,
        12: 2,
        13: 0,
        14: 1,
        15: 0
    }
    for line in open("inputB.txt"):
        s = line.split()
        if linec == 0:
            linec += 1
            continue
        if linec % 2:
            n = int(s[0])
            linec += 1
        else:
            linec += 1
            arr = [int(x) for x in s]
            num_factors = [fot[x] for x in arr]
            tot_factors = sum(num_factors)
            pr = [el / 2 ** num_factors[i] for i, el in enumerate(arr)]
            base = sum(pr)
            max_res = sum(arr)
            for i, a in enumerate(arr):
                max_res = max(a * 2 ** (tot_factors - num_factors[i]) + base - a // 2 ** num_factors[i], max_res)
            print(int(max_res))
problemB()
```

```python
def problemC():
    linec = 0
    for line in open("inputC.txt"):
        s = line.split()
        if linec == 0:
            linec += 1
            continue
        if linec % 2:
            linec += 1
            n = int(s[0])
        else:
            linec += 1
            arr = [int(x) for x in s]
            max_res = 0
            for rot in range(0, len(arr)):
                arr = arr[rot:] + arr[:rot]
                stack = []
                res = 0
                for el in arr:
                    if len(stack) < 2:
                        stack.append(el)
                    else:
                        if stack[-2] == el:
                            stack.append(el)
                        else:
                            while(len(stack) >= 2 and stack[-2] != el):
                                stack.pop()
                                res += 1
                            stack.append(el)
                res += len(stack) // 2 + 1
                #print(f"stack: { stack }")
                # print(f"res: { res }")
                max_res = max(res, max_res)
            print(max_res)
problemC()
```

```python
# Reference: https://www.geeksforgeeks.org/read-a-file-line-by-line-in-python/

def problemD():
    file = open("inputD.txt", "r")
    N = int(file.readline().split()[0])
    for i in range(N):
        line = file.readline().split()
        n, q = int(line[0]), int(line[1])
        arr = sorted([int(x) for x in file.readline().split()])[::-1]
        a = 0
        presum = []
        for el in arr:
            a += el
            presum.append(a)
        for j in range(q):
            query = int(file.readline().split()[0])
            res = binary_search(presum, query)
            if res != None: print(res + 1)
            else: print(-1)


def binary_search(presum, target):
    left = 0
    right = len(presum) - 1
    res = None
    while left <= right:
        mid = (left + right) // 2
        if presum[mid] >= target:
            res = mid
            right = mid - 1
        else:
            left = mid + 1
    return res
problemD()
```

```python
def problemE():
    file = open("inputE.txt", "r")
    N = int(file.readline().split()[0])
    for i in range(N):
        n = int(file.readline().split()[0])
        baseline = sum([get_factors_2(int(x)) for x in file.readline().split()])
        possibilities = sorted([get_factors_2(x) for x in range(1, n + 1)])[::-1]

        if baseline >= n:
            print(0)
        elif baseline + sum(possibilities) < n:
            print(-1)
        else:
            curr = baseline
            for i, el in enumerate(possibilities):
                curr += el
                if curr >= n: break
            print(i + 1)
        
def get_factors_2(num):
    res = 0
    while not (num % 2):
        num //= 2
        res += 1
    return res
problemE()
```
