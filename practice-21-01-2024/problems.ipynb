```python
def problemB():

    file = open("inputB.txt")
    read = file.readline

    all_numbers = []
    towers = []
    n = int(read().split()[0])

    for _ in range(n):
        t = [int(x) for x in read().split()][1:]
        all_numbers += t
        towers.append(t)
    
    all_numbers = sorted(all_numbers)
    index_of = {}
    for i, el in enumerate(all_numbers):
        index_of[el] = i

    
    num_splits = 0
    for t in towers:
        for i in range(len(t) - 1):
            next_index = index_of[t[i]] + 1
            if next_index >= len(all_numbers):
                num_splits += 1
                continue
            if t[i + 1] != all_numbers[next_index]:
                num_splits += 1
    
    n += num_splits
    print(f"{ num_splits } { n - 1 }")

problemB()
```

```python
class UnionFind:
    def __init__(self, nodes):
        self.reps = {}
        for el in nodes:
            self.reps[el] = el
        
    def find(self, node):
        if self.reps[node] == node: return self.reps[node]
        r = self.find(self.reps[node])
        self.reps[node] = r
        return r
    
    def merge(self, a, b):
        rep_a = self.find(a)
        rep_b = self.find(b)
        if rep_a > rep_b:
            self.reps[rep_b] = rep_a 
        else:
            self.reps[rep_a] = rep_b
        
def problemD():

    file = open("inputD.txt")
    read = file.readline
    
    m, n = [int(x) for x in read().split()]
    a = [int(x) for x in read().split()]
    b = [int(x) for x in read().split()]

    edges = []
    no = set()
    for i in range(m):
        s = [int(x) for x in read().split()[1:]]
        if len(s) == 1: continue
        for j, el in enumerate(s):
            edges.append([a[i] + b[el - 1], i + 1, el])
            no.add(el)
    nodes = []
    for el in no:
        nodes.append(el + m)
    for el in range(1, m + 1):
        nodes.append(el)
    
    Uf = UnionFind(nodes)
    edges = sorted(edges, key=lambda x: -x[0])
    
    res = 0
    for cost, a, b in edges:
        if Uf.find(a) == Uf.find(b + m): 
            res += cost
        else:
            Uf.merge(a, b + m)

    print(res)

problemD()
```
